```python
import pandas as pd


train=pd.read_csv('train.csv')
test=pd.read_csv('test.csv')
ideal=pd.read_csv('ideal.csv')
```

```python
from sqlalchemy import create_engine,Column,MetaData,Table,Float
```

```python
db_name='test'
engine=create_engine('sqlite:///%s.db'% db_name,echo=False)
```

```python
from sklearn.metrics import mean_squared_error

tr=train.drop('x',axis=1)
id=ideal.drop('x',axis=1)
min_result={}

for i in tr.items():
    r={}
    for j in id.items():
        
        mse= mean_squared_error(i[1],j[1])
        
        r[j[0]]=mse


    min_result[i[0]]=(min(r, key=r.get),min(r.values())) 
id_list=['x']+[i[0] for i in list(min_result.values())]
ideal= ideal[id_list]


testdev= pd.merge(test, train, on='x', how='inner')


# deviation between y1 train and equivalent y from ideal
maxdev=[]
temp=0
for col in range(1,5):
    for row in range(0,len(train)):
        if abs(train.iloc[row,col]- ideal.iloc[row,col]) > temp:
            temp=abs(train.iloc[row,col]- ideal.iloc[row,col])
            
    maxdev.append(temp)


lists=[]
dev=0
for col in range(2,6):
    for row in range(0,len(testdev)):
        if all(abs(testdev.iloc[row,1]- testdev.iloc[row,col])<maxdev):
            dev= abs(testdev.iloc[row,1]- testdev.iloc[row,col])
            lists.append((test.iloc[row]['x'],test.iloc[row]['y'],dev,maxdev)) #save the row to list
            
            
    
savedtest= pd.DataFrame(data=lists, columns=['X','Y','dev','maxdev'])
```

```python
savedtest=savedtest.applymap(str)
savedtest.to_sql(con=engine, name='test', if_exists='replace', index=False)
```
